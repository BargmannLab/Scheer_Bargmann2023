```python
#Minutes roaming/dwelling per leaving event light OFF/ON
from scipy import stats

def calculate_ci_binom(sample, numMins, confidence = 0.95): #standard error of a count (poisson)
    n = sample.sum() # number of lawn leaving events in the sample
    p = n/numMins # the rate of successes (lawn leaving)
    q = 1-p #the rate of failures (no leaving)
    norm_test_stat = stats.norm.ppf((confidence + 1)/2)
    confInt = (norm_test_stat * np.sqrt(n*p*q))/numMins #scale it to the number of minutes in the intervals in question (i.e. total time roaming or dwelling)
    return confInt

#https://stats.stackexchange.com/questions/15371/how-to-calculate-a-confidence-level-for-a-poisson-distribution
def calculate_ci_poisson(sample, numMins, confidence = 0.95): #standard error of a count (poisson)
    n = sample.count()
    rate = sample.sum()/numMins # number of lawn leaving events per minute per animal
    norm_test_stat = stats.norm.ppf((confidence + 1)/2)
    confInt = (norm_test_stat * np.sqrt(rate/n)) #scale it to the number of minutes in the intervals in question (i.e. total time roaming or dwelling)
    return confInt


binLight = binData(LIGHT_STIM,binSize,np.nanmean).astype(bool)[0]
# lightOnIdx = np.where(binLight)[0]
# lightOffIdx = np.where(~binLight)[0]

#light OFF = min 0-10, 22-30; light ON = min 12-20, 32-40
lightOffIdx = np.hstack((np.arange(0,10*6),np.arange(22*6,30*6)))
lightOnIdx = np.hstack((np.arange(12*6,20*6),np.arange(32*6,40*6)))

# #just first OFF and ON:
# lightOffIdx = np.arange(0,10*6)
# lightOnIdx = np.arange(12*6,20*6)

data_list = []
for d_key,d_val in Data_subs.items():
    numAssays = d_val['matfilename'].shape[0]
    
    LL_lightOFF = d_val['bin_LawnExit_mostRecent'][:,lightOffIdx]
    RD_lightOFF = d_val['RD_states_Matrix_exog_Cent'][:,lightOffIdx]
    InLawn_lightOFF = d_val['InLawnRunMask'][:,lightOffIdx]
    numLLevents_lightOFF = np.sum(np.sum(LL_lightOFF))
    fracRoam_lightOFF = np.sum(np.sum(RD_lightOFF))/np.sum(np.sum(InLawn_lightOFF))
    numMinRoam_lightOFF = np.sum(np.sum(RD_lightOFF==True))/6
    numMinDwell_lightOFF = np.sum(np.sum(RD_lightOFF==False))/6
    numLL_Roam_lightOFF = np.sum(np.sum(np.logical_and(LL_lightOFF,RD_lightOFF==True)))
    numLL_Dwell_lightOFF = np.sum(np.sum(np.logical_and(LL_lightOFF,RD_lightOFF==False)))
    LLperMin_Roam_lightOFF = numLL_Roam_lightOFF/numMinRoam_lightOFF
    LLperMin_Dwell_lightOFF = numLL_Dwell_lightOFF/numMinDwell_lightOFF
    MinperLL_Roam_lightOFF = numMinRoam_lightOFF/numLL_Roam_lightOFF
    MinperLL_Dwell_lightOFF = numMinDwell_lightOFF/numLL_Dwell_lightOFF
    
    leavingWhileRoaming = np.sum(np.logical_and(LL_lightOFF,RD_lightOFF==True),axis=1)
    df = pd.DataFrame({"leavingWhileRoaming":leavingWhileRoaming})
    lwR_ci_lightOFF = calculate_ci_poisson(df.leavingWhileRoaming, numMinRoam_lightOFF, confidence = 0.95)
    leavingWhileDwelling = np.sum(np.logical_and(LL_lightOFF,RD_lightOFF==False),axis=1)
    df = pd.DataFrame({"leavingWhileDwelling":leavingWhileDwelling})
    lwD_ci_lightOFF = calculate_ci_poisson(df.leavingWhileDwelling, numMinDwell_lightOFF, confidence = 0.95)
    
    LL_lightON = d_val['bin_LawnExit_mostRecent'][:,lightOnIdx]
    RD_lightON = d_val['RD_states_Matrix_exog_Cent'][:,lightOnIdx]
    InLawn_lightON = d_val['InLawnRunMask'][:,lightOnIdx]
    numLLevents_lightON = np.sum(np.sum(LL_lightON))
    fracRoam_lightON = np.sum(np.sum(RD_lightON))/np.sum(np.sum(InLawn_lightON))
    numMinRoam_lightON = np.sum(np.sum(RD_lightON==True))/6
    numMinDwell_lightON = np.sum(np.sum(RD_lightON==False))/6
    numLL_Roam_lightON = np.sum(np.sum(np.logical_and(LL_lightON,RD_lightON==True)))
    numLL_Dwell_lightON = np.sum(np.sum(np.logical_and(LL_lightON,RD_lightON==False)))
    LLperMin_Roam_lightON = numLL_Roam_lightON/numMinRoam_lightON
    LLperMin_Dwell_lightON = numLL_Dwell_lightON/numMinDwell_lightON
    MinperLL_Roam_lightON = numMinRoam_lightON/numLL_Roam_lightON
    MinperLL_Dwell_lightON = numMinDwell_lightON/numLL_Dwell_lightON
    
    leavingWhileRoaming = np.sum(np.logical_and(LL_lightON,RD_lightON==True),axis=1)
    df = pd.DataFrame({"leavingWhileRoaming":leavingWhileRoaming})
    lwR_ci_lightON = calculate_ci_poisson(df.leavingWhileRoaming, numMinRoam_lightON, confidence = 0.95)
    leavingWhileDwelling = np.sum(np.logical_and(LL_lightON,RD_lightON==False),axis=1)
    df = pd.DataFrame({"leavingWhileDwelling":leavingWhileDwelling})
    lwD_ci_lightON = calculate_ci_poisson(df.leavingWhileDwelling, numMinDwell_lightON, confidence = 0.95)
    
    df = pd.DataFrame({'# assays': numAssays,
                                        '# events light OFF': numLL_Dwell_lightOFF+numLL_Roam_lightOFF,
                                       '# events light ON': numLL_Dwell_lightON+numLL_Roam_lightON,
                                       'fraction of time roaming light OFF': fracRoam_lightOFF,
                                       'fraction of time roaming light ON': fracRoam_lightON,
                                       '# mins roaming light OFF': numMinRoam_lightOFF,
                                       '# mins roaming light ON': numMinRoam_lightON,
                                       '# mins dwelling light OFF': numMinDwell_lightOFF,
                                       '# mins dwelling light ON': numMinDwell_lightON,
                                       '# events while roaming light OFF': numLL_Roam_lightOFF,
                                       '# events while roaming light ON': numLL_Roam_lightON,
                                      '# events while dwelling light OFF': numLL_Dwell_lightOFF,
                                       '# events while dwelling light ON': numLL_Dwell_lightON,
                                       'roam: events/min light OFF': LLperMin_Roam_lightOFF,
                                       'roam: events/min light ON': LLperMin_Roam_lightON,
                                       'dwell: events/min light OFF': LLperMin_Dwell_lightOFF,
                                       'dwell: events/min light ON': LLperMin_Dwell_lightON,
                                       'roam: min/event light OFF': MinperLL_Roam_lightOFF,
                                       'roam: min/event light ON': MinperLL_Roam_lightON, 
                                       'dwell: min/event light OFF': MinperLL_Dwell_lightOFF,
                                        'dwell: min/event light ON': MinperLL_Dwell_lightON,
                                      'roam: e/m poisson ci_sym light OFF': lwR_ci_lightOFF,
                                      'dwell: e/m poisson ci_sym light OFF': lwD_ci_lightOFF,
                                      'roam: e/m poisson ci_sym light ON': lwR_ci_lightON,
                                      'dwell: e/m poisson ci_sym light ON':lwD_ci_lightON}, index=[d_key])
                    

    data_list.append(df)
    
roam_dwell_per_leave_DF = pd.concat(data_list)
```

```python
#Perform a Kolmogorov-Smirnov 2sample test across all genotype pairs, and then Bonferroni correct it
from scipy.stats import ks_2samp

#first LL light ON
firstLL_lightON_pvals = []
for i, p in enumerate(pairs):
    _,pval = ks_2samp(first_LL_lightON_acrossgen[p[0]], first_LL_lightON_acrossgen[p[1]], alternative='two-sided')
    firstLL_lightON_pvals.append(pval)
firstLL_lightON_pvals = np.array(firstLL_lightON_pvals)*len(pairs) #Bonferroni correct

#first LL light OFF
firstLL_lightOFF_pvals = []
for i, p in enumerate(pairs):
    _,pval = ks_2samp(first_LL_lightOFF_acrossgen[p[0]], first_LL_lightOFF_acrossgen[p[1]], alternative='two-sided')
    firstLL_lightOFF_pvals.append(pval)
firstLL_lightOFF_pvals = np.array(firstLL_lightOFF_pvals)*len(pairs) #Bonferroni correct

#first HP light ON
firstHP_lightON_pvals = []
for i, p in enumerate(pairs):
    _,pval = ks_2samp(first_HP_lightON_acrossgen[p[0]], first_HP_lightON_acrossgen[p[1]], alternative='two-sided')
    firstHP_lightON_pvals.append(pval)
firstHP_lightON_pvals = np.array(firstHP_lightON_pvals)*len(pairs) #Bonferroni correct

#first HP light OFF
firstHP_lightOFF_pvals = []
for i, p in enumerate(pairs):
    _,pval = ks_2samp(first_HP_lightOFF_acrossgen[p[0]], first_HP_lightOFF_acrossgen[p[1]], alternative='two-sided')
    firstHP_lightOFF_pvals.append(pval)
firstHP_lightOFF_pvals = np.array(firstHP_lightOFF_pvals)*len(pairs) #Bonferroni correct



#print pairs and significance stars to a text file
with open(subfolder+'/firstEventTiming_pVals.txt', 'w') as f:
    f.write('first LL event, light ON:')
    f.write('\n')
    for pair, p in zip(pairs,firstLL_lightON_pvals):
        f.write(str(pair)+','+str(p)+','+str(categorizePValue(p)))
        f.write('\n')
    f.write('first LL event, light OFF:')
    f.write('\n')
    for pair, p in zip(pairs,firstLL_lightOFF_pvals):
        f.write(str(pair)+','+str(p)+','+str(categorizePValue(p)))
        f.write('\n')
    f.write('first HP event, light ON:')
    f.write('\n')
    for pair, p in zip(pairs,firstHP_lightON_pvals):
        f.write(str(pair)+','+str(p)+','+str(categorizePValue(p)))
        f.write('\n')
    f.write('first HP event, light OFF:')
    f.write('\n')
    for pair, p in zip(pairs,firstHP_lightOFF_pvals):
        f.write(str(pair)+','+str(p)+','+str(categorizePValue(p)))
        f.write('\n')
    f.close()
```
